```python
import time

all_signatures = []
oldest_sig = None
max_pages = 50 # 1000 sigs per page, so max 50,000 transactions to save API credits

print(f"Backwards travel to the creation of {token_address[:8]}...\n")

for page in range(max_pages):
    payload = {
        "jsonrpc": "2.0",
        "id": 1,
        "method": "getSignaturesForAddress",
        "params": [
            token_address,
            {"limit": 1000}
        ]
    }
    
    # If not the first page, tell it to look BEFORE the oldest signature we have!
    if oldest_sig:
        payload["params"][1]["before"] = oldest_sig
        
    response = requests.post(rpc_url, json=payload).json()
    batch = response.get('result', [])
    
    if not batch:
        print(f"Reached the very first block! (Total Transactions: {len(all_signatures)})")
        break
        
    sigs = [tx['signature'] for tx in batch]
    all_signatures.extend(sigs)
    oldest_sig = sigs[-1] # The last one in the list is the oldest timestamp
    
    print(f"Page {page+1}: Found {len(sigs)} signatures. Oldest passing block: {oldest_sig[:15]}...")
    time.sleep(0.1)

# Because getSignaturesForAddress returns from Newest to Oldest,
# We must reverse the list so our Graph is built Chronologically (Creation -> Dev Activity -> Rug Pull)
chronological_signatures = all_signatures[::-1]

print(f"Total Timeline Built: {len(chronological_signatures)} transactions!")
if len(chronological_signatures) > 0:
    print(f"Token Created (Origin Block): {chronological_signatures[0]}")
    print(f"Latest Activity: {chronological_signatures[-1]}")
```

```python
import pandas as pd
import requests
import time
import os
from dotenv import load_dotenv

# Load Helius RPC
load_dotenv(dotenv_path='../.env')
rpc_url = f"https://mainnet.helius-rpc.com/?api-key={os.getenv('HELIUS_API_KEY')}"

# Load the SolRPDS Scam Dataset you downloaded earlier
dataset_path = "../exported_data/SolRPDS_2023.csv"
print(f"Loading Dataset: {dataset_path}...")
df_scams = pd.read_csv(dataset_path)

# Let's run a test extraction on the first 3 Scam Tokens
test_tokens = df_scams['MINT'].dropna().unique()[:200]
print(f"Starting extraction pipeline for {len(test_tokens)} tokens...\n")

master_edges = []

for idx, token in enumerate(test_tokens):
    print(f"[{idx+1}/{len(test_tokens)}] Crawling Token: {token}")
    
    # A. Crawl backwards in time (Max 5 pages / 5,000 blocks to keep test fast) --
    signatures = []
    oldest_sig = None
    for page in range(5):
        payload = {"jsonrpc": "2.0", "id": 1, "method": "getSignaturesForAddress", "params": [token, {"limit": 1000}]}
        if oldest_sig: payload["params"][1]["before"] = oldest_sig
        
        res = requests.post(rpc_url, json=payload).json()
        batch = res.get('result', [])
        if not batch: break
        
        signatures.extend([tx['signature'] for tx in batch])
        oldest_sig = signatures[-1]
        time.sleep(0.1) # Safety timeout
        
    chronological = signatures[::-1]
    print(f"  -> Found {len(chronological)} raw transactions. Extracting Edges...")
    
    # B. Edge Extraction (first 300 blocks of the tokens lifespan)
    for sig in chronological[:300]:
        tx_payload = {"jsonrpc": "2.0", "id": 1, "method": "getTransaction", "params": [sig, {"encoding": "jsonParsed", "maxSupportedTransactionVersion": 0}]}
        tx_res = requests.post(rpc_url, json=tx_payload).json()
        
        if 'result' not in tx_res or not tx_res['result']: continue
        meta = tx_res['result'].get('meta')
        if not meta or meta.get('err'): continue
            
        wallet_deltas = {}
        for bal in meta.get('preTokenBalances', []):
            if bal['mint'] == token: wallet_deltas.setdefault(bal.get('owner'), {"pre": 0, "post": 0})["pre"] = float(bal['uiTokenAmount']['uiAmount'] or 0)
        for bal in meta.get('postTokenBalances', []):
            if bal['mint'] == token: wallet_deltas.setdefault(bal.get('owner'), {"pre": 0, "post": 0})["post"] = float(bal['uiTokenAmount']['uiAmount'] or 0)
            
        senders, receivers = [], []
        for owner, balances in wallet_deltas.items():
            delta = balances["post"] - balances["pre"]
            if delta > 0: receivers.append((owner, delta))
            elif delta < 0: senders.append((owner, abs(delta)))
                
        for s, s_amt in senders:
            for r, r_amt in receivers:
                amt = min(s_amt, r_amt)
                if amt > 0:
                    master_edges.append({
                        "Token": token,
                        "Signature": sig,
                        "Sender": s,
                        "Receiver": r,
                        "Amount(token_units)": amt
                    })
        time.sleep(0.1)

# 3. Export to our new Master Edges Dataset!
edges_df = pd.DataFrame(master_edges)
edges_df.to_csv("../exported_data/graph_edges_dataset.csv", index=False)

print(f"Phase 1 Pipeline Complete!")
print(f"Saved {len(edges_df)} Relational Edges to '../exported_data/graph_edges_dataset.csv'")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import time

# Initialize the Directed Graph
G = nx.DiGraph()

print("Extracting Edges and Building Graph... (Taking the first 50 blocks)")

for i, sig in enumerate(chronological_signatures[:1000]):
    tx_payload = {"jsonrpc": "2.0", "id": 1, "method": "getTransaction", "params": [sig, {"encoding": "jsonParsed", "maxSupportedTransactionVersion": 0}]}
    tx_res = requests.post(rpc_url, json=tx_payload).json()
    
    if 'result' not in tx_res or tx_res['result'] is None: continue
    meta = tx_res['result']['meta']
    if not meta or meta.get('err'): continue
        
    wallet_deltas = {}
    for bal in meta.get('preTokenBalances', []):
        if bal['mint'] == token_address:
            wallet_deltas.setdefault(bal.get('owner'), {"pre": 0, "post": 0})["pre"] = float(bal['uiTokenAmount']['uiAmount'] or 0)
            
    for bal in meta.get('postTokenBalances', []):
        if bal['mint'] == token_address:
            wallet_deltas.setdefault(bal.get('owner'), {"pre": 0, "post": 0})["post"] = float(bal['uiTokenAmount']['uiAmount'] or 0)
            
    senders = []
    receivers = []
    
    for owner, balances in wallet_deltas.items():
        delta = balances["post"] - balances["pre"]
        if delta > 0: receivers.append((owner, delta))
        elif delta < 0: senders.append((owner, abs(delta)))
            
    if senders and receivers:
        for sender, send_amt in senders:
            for receiver, rec_amt in receivers:
                amt = min(send_amt, rec_amt)
                if amt > 0:
                    
                    # edge into graph 
                    if G.has_edge(sender, receiver):
                        G[sender][receiver]['weight'] += amt # Add volume if they trade again
                    else:
                        G.add_edge(sender, receiver, weight=amt)
                        
    time.sleep(0.1)

print(f"\nGraph Built! Total Unique Wallets (Nodes): {G.number_of_nodes()} | Total Flows (Edges): {G.number_of_edges()}")

# Netowrk Topology 
plt.figure(figsize=(10, 10))

# The spring layout naturally pushes connected nodes together and pushes unconnected nodes apart (like bubbles!)
pos = nx.spring_layout(G, k=0.5, iterations=50)

# Make wallets that sent/received a lot of transactions visually larger!
degrees = dict(G.degree())
node_sizes = [v * 150 for v in degrees.values()]

nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='cyan', alpha=0.8, edgecolors='black')
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=15, edge_color='gray', alpha=0.6)

plt.title("Token Transfer Graph Topology", fontsize=16)
plt.axis("off")
plt.show()
```

```python
import ast

def get_risk_level(audit_string):
    # Some rows might be missing data (NaN)
    if pd.isna(audit_string):
        return "unknown"
        
    try:
        # ast.literal_eval safely converts the text "{'riskLevel': 'high'}" into an actual Python dictionary
        audit_dict = ast.literal_eval(audit_string) 
        
        # safely extract riskLevel without throwing an error if it's missing
        return audit_dict.get('riskLevel', 'unknown') 
    except Exception as e:
        return "error"

# create a new risk_level column 
safe_tokens['Risk_Level'] = safe_tokens['audit_data'].apply(get_risk_level)

# check results
safe_tokens["Risk_Level"].value_counts()

# Uknown is probably that the auditor AI wasnt able fast enough to track and audit a newly released token I fed it , ignore
# For the sake of this exercise, we will sample some "low" risk_level tokens 

safe_tokens_df = safe_tokens.loc[safe_tokens["Risk_Level"] == "low"]
# typical trenches names lmao 
safe_tokens_df.head(10)
```
